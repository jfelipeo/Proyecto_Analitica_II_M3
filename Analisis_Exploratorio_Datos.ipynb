## Duplicados

```python
print("Datos con duplicados:",datos.duplicated().sum(), " // ",datos.shape)

# Eliminar duplicados
datos = datos.drop_duplicates()

print ("Datos sin duplicados:",datos.shape)
```

## Datos Atípicos

```python
# Seleccionar las columnas numéricas excluyendo 'ID'
columnas_numericas = datos.select_dtypes(include=['int64', 'float64']).drop(columns=['ID'])

# Configurar el tamaño de la figura
plt.figure(figsize=(20, 25))

# Crear los diagramas de caja
for i, columna in enumerate(columnas_numericas.columns, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(data=datos[columna], orient='vertical', width=0.5)
    plt.title(columna)

# Ajustar el espacio entre gráficos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Variables de interés
variables_interes = ["Service time", "Age", "Work load Average/day ", "Hit target", "Disciplinary failure", "Education", "Pet", "Height", "Absenteeism time in hours"]

# Diccionario para almacenar los resultados
conteo_valores_extremos = {}

# Calcular el rango intercuartílico (IQR) para cada variable y contar valores extremos
for variable in variables_interes:
    q1 = datos[variable].quantile(0.25)
    q3 = datos[variable].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    
    # Contar valores extremos
    outliers_count = len(datos[(datos[variable] < lower_bound) | (datos[variable] > upper_bound)])
    
    # Almacenar los resultados en el diccionario
    conteo_valores_extremos[variable] = outliers_count

# Crear un DataFrame a partir del diccionario de resultados
df_valores_extremos = pd.DataFrame.from_dict(conteo_valores_extremos, orient='index', columns=['Conteo de valores extremos'])

# Mostrar la tabla de conteo de valores extremos
print(df_valores_extremos)
```
