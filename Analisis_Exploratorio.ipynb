## Escalado de variables numéricas

```python
# Copiar el DataFrame
df1 = datos.copy()
```

```python
sca_cols = df1[['Transportation expense','Distance from Residence to Work','Service time','Age','Work load Average/day ','Hit target','Weight','Height','Absenteeism time in hours']]

scaler = StandardScaler()
scaled_df1 = pd.DataFrame(scaler.fit_transform(sca_cols), columns=sca_cols.columns)
scaled_df1.head()
```

```python
# Reestablecer los índices en ambos dataset para unirlos
scaled_df1.reset_index(drop=True, inplace=True)
df1.reset_index(drop=True, inplace=True)
```

```python
# Union de las variables escaladas a las demas columnas

cols = ['Transportation expense','Distance from Residence to Work','Service time','Age','Work load Average/day ','Hit target','Weight','Height','Absenteeism time in hours']

# Convertir las columnas seleccionadas en df1 a tipo float64
df1[cols] = df1[cols].astype(float)

df1[cols]=scaled_df1
df1.head()
```

## Datos Atípicos

```python
# Configurar el tamaño de la figura
plt.figure(figsize=(20, 25))

# Crear los diagramas de caja
for i, columna in enumerate(df1.columns, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(data=df1[columna], orient='vertical', width=0.5)
    plt.title(columna)

# Ajustar el espacio entre gráficos
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

### Conteo de datos atípicos extremos

```python
# Diccionario de valores límite para cada columna
valores_limite = {'Work load Average/day ': 2, 'Height': 3, 'Absenteeism time in hours': 6}

# Conteo de datos
conteos = {}

# Iteraracion y busqueda
for columna, limite in valores_limite.items():
    # Filtrar los datos de la columna que sean mayores o iguales al valor límite
    datos_filtrados = df1[df1[columna] >= limite][columna]
    # Contar el número de filas encontradas
    conteo = datos_filtrados.count()
    # Almacenar el conteo en el diccionario
    conteos[columna] = conteo

# Mostrar los conteos para cada columna
for columna, conteo in conteos.items():
    print("Datos mayores o iguales a", valores_limite[columna], "en la columna", columna + ":", conteo)
```

### Eliminación de datos atípicos

```python
# Tomamos los indeces de los datos que superan el valor limite
outs_idx=[df1[df1['Absenteeism time in hours'] >= 6]['Absenteeism time in hours'].index.astype(int)]

# Seleccionamos valores unicos para los indices

outs_idx=pd.Series(sorted(np.concatenate(outs_idx))).unique()

# Eliminamos los datos que superan el valor limite

df1_out = df1.drop(index=outs_idx)
df1_out.reset_index(drop=True, inplace=True)

print(df1_out.shape)
```

## Codificacion de variables con la frecuencia en cada categoría

```python
df1_out_frec = df1_out.copy()
# Columnas para calcular frecuencia
columns_to_encode = ['Reason for absence']

# Calcular frecuencia y aplicar codificación para cada columna
for column in columns_to_encode:
    freq = df1_out_frec[column].value_counts(normalize=True)
    df1_out_frec[column] = df1_out_frec[column].map(freq)
```
